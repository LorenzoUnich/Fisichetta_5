```python
def exp(x,a,b,c):
    return np.exp(-x/b)*a+c
def frontselect(lista, fronte=0):
    result = []
    for i in lista:
        if i[2]==fronte:
            result.append(i)
    return result
def convtime(x, a=1.044, b= -159):
    return a*x+b
def fromhist(x):
    l = len(x)
    xnew = []
    for i in range(l-1):
        xnew.append((x[i]+x[i+1])/2)
    return xnew
def first_zero(y):
    l = len(y)
    re = -1
    i = 0
    while i<l:

        if y[i] == 0:
            
            re = i
            break
        i = i+1
    
    #if there is no zero, the function returns -1
    if re ==-1:
        print("There's no zero in the list!")
    return re
        
```
