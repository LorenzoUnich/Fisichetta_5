```python
from tdctofile import tdc_to_events as tte
import numpy as np
import matplotlib.pyplot as plt
from fit_pesato_3 import fit 


```

### Common stop


```python
tempi = [306,324,372,566,620,710,750,840,930,970,1920]
mediet = []
stdt= []
errt = []
for i in tempi[2:]:
    eventi = tte(str(i)+"ns.txt")[1:]
    ts = [j[3] for j in eventi]
    mediet.append(np.mean(ts))
    s = np.std(ts)
    stdt.append(s)
    errt.append(s/(np.sqrt(len(eventi))))
    
    
```

```python
len(tempi)
```

```python
def retta(a,b,x):
    return a+x*b
```

```python
plt.figure(figsize= (6,4), dpi = 100)
plt.scatter(tempi[2:],mediet)
f = fit(tempi[2:],mediet,errt)
plt.plot(tempi,[retta(f[0],f[1],i) for i in tempi], color = "r")
```
